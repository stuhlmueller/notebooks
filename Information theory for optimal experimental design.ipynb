### Prelude

```python
from __future__ import division
from random import random
from collections import defaultdict
from math import log

def plot_bars(xs, ys, ylab, title):
    fig = pylab.figure()
    ax = fig.add_subplot(1,1,1)
    N = len(ys)
    ind = range(N)
    ax.bar(ind, ys, facecolor='#777777', align='center')
    ax.set_ylabel(ylab)
    ax.set_title(title,fontstyle='italic')
    ax.set_xticks(ind)
    group_labels = xs
    ax.set_xticklabels(group_labels)
    fig.autofmt_xdate()
    pylab.show()
    
def normalize(counts):
    Z = sum(counts.values())
    return dict((k, c/Z) for (k, c) in counts.items())

def dist(model, n=10000):
    counts = defaultdict(lambda: 0)
    for _ in range(n):
        sample = model()
        counts[sample] += 1
    return normalize(counts)

def plot_model(model):
    d = dist(model)
    return plot_bars(d.keys(), d.values(), 
                     ylab='Probability',
                     title='Probability by outcome (%s)' % model.__name__)   
```

## Information theory for optimal experimental design

### Random choice

```python
def flip(p):
    return random() < p

def fair_coin():
    return flip(.77)

plot_model(fair_coin)
```

### A simple model

```python
def gaming_causes_cognition():
    plays_game = flip(.3)
    high_cognition = flip(.9) if plays_game else flip(.1)
    return plays_game, high_cognition

plot_model(gaming_causes_cognition)
```

### Conditioning

```python
def condition(model, predicate):
    def new_model():
        sample = None
        while not sample or not predicate(sample):
            sample = model()
        return sample
    return new_model

def has_high_cognition(vals):
    plays_game, high_cognition = vals
    return high_cognition

plot_model(gaming_causes_cognition)
plot_model(condition(gaming_causes_cognition, has_high_cognition))
```

### Expectations

$$ E_{p}[f(x)] = \sum_x p(x) f(x) $$

```python
def expectation(model, f):
    ex = 0.0
    for (val, prob) in dist(model).items():
        ex += prob * f(val, prob)
    return ex

def unfair_coin():
    return flip(.9)

def f(v, prob):
    return 1 if v else 0

expectation(unfair_coin, f)
```

### Entropy

$$ H(X) = \sum_x p(x) \log p(x) $$

```python
def surprisal(prob):
    return -log2(prob)

def entropy(model):
    return expectation(model, lambda val, prob: surprisal(prob))

print entropy(fair_coin)
print entropy(unfair_coin)
```

### Projection

```python
def project(model, index):
    def marginal_model():
        sample = model()
        return sample[index]
    return marginal_model

plot_model(project(gaming_causes_cognition, 0))
```

### Intervention

```python
def gaming_causes_cognition_fixcog():
    plays_game = flip(.3)
    high_cognition = True
    return plays_game, high_cognition

plot_model(gaming_causes_cognition_fixcog)
```

### A hierarchical model

```python
def gaming_causes_cognition():
    plays_game = flip(.3)
    high_cognition = flip(.9) if plays_game else flip(.1)
    return plays_game, high_cognition

def cognition_causes_gaming():
    high_cognition = flip(.3)
    plays_game = flip(.9) if high_cognition else flip(.1)
    return plays_game, high_cognition

def hierarchical_model():
    if flip(.55):
        hypothesis = "gaming_causes_cognition"
    else:
        hypothesis = "cognition_causes_gaming"
    plays_game, high_cognition = eval(hypothesis)()
    return hypothesis, plays_game, high_cognition

plot_model(gaming_causes_cognition)
plot_model(cognition_causes_gaming)
plot_model(hierarchical_model)
```

### When observations are uninformative...

```python
def observe_subject(vals):
    hypothesis, plays_game, high_cognition = vals
    return (plays_game == True) and (high_cognition == True)

plot_model(project(hierarchical_model, 0))
plot_model(project(condition(hierarchical_model, observe_subject), 0))
```

### ...Interventions can help

```python
def observe_subject(vals):
    hypothesis, plays_game, high_cognition = vals
    return (plays_game == True) and (high_cognition == True)

plot_model(project(hierarchical_model_fixgame, 0))
plot_model(project(condition(hierarchical_model_fixgame, observe_subject), 0))

print entropy(project(hierarchical_model_fixgame, 0))
print entropy(project(condition(hierarchical_model_fixgame, observe_subject), 0))
```

### Expected entropy after an intervention

```python
def expected_posterior_entropy(model):
    def posterior_entropy(cog, prob):
        return entropy(project(condition(model, lambda args: args[2] == cog), 0))
    return expectation(project(model, 2), posterior_entropy)

print expected_posterior_entropy(hierarchical_model_fixcog)
```
